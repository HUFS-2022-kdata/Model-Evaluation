```python
def show_directory(path):
    transcript_tmp = []
    path += '/'
    for folder in os.listdir(PATH):
       for filename in os.listdir(os.path.join(PATH,folder)):
          # filename = os.path.join(path, file)
          ext = os.path.splitext(filename)[-1]
          if ext == '.txt':
              f = open(os.path.join(PATH,folder,filename),"r",encoding='cp949')
              transcript = f.read()
              transcript_tmp.append(transcript)
    return transcript_tmp
```

```python
def Recog(PATH):
    r = sr.Recognizer()
    quest_name_list = []
    transcript_tmp = []
    answer_stt = []
    for folder in os.listdir(PATH):
       for file in os.listdir(os.path.join(PATH,folder)):
          if ".txt" in file:
              with open(os.path.join(PATH,folder,file), "r", encoding='cp949') as f:
                
                  key = re.sub('.+?(문제\d+).txt','\1',file)
                  key = key.split('.')[0]
                  quest_name_list.append(key)
                  answer_voice = PATH +"/"+ folder + "/" + key + ".wav"
              with sr.AudioFile(answer_voice) as source:
                  try:  
                    audio = r.record(source, duration=120)
                    answer_stt.append(r.recognize_google(audio_data=audio, language='ko-KR'))
                  except Exception as e:
                    print("Error : ")
                    print(file)
    return quest_name_list, answer_stt
```

```python
script_info = show_directory(PATH)
```

```python
quest_name_list, answer_stt = Recog(PATH)
```

# 문장 유사도 검증

## 1)코사인 유사도 비교

```python
def cos_sim(a, b):
    return dot(a, b)/(norm(a)*norm(b))
```

```python
def make_matrix(feats, list_data):
    freq_list = []
    for feat in feats:
        freq = 0
        for word in list_data:
            if feat == word:
                freq += 1
        freq_list.append(freq)
    return freq_list
```

```python
okt = Okt() # 질문: 과연 문장의 일치 여부를 조사하기 위해 형태소 분석기를 활용하는 게 맞나? 
i = 0
cos_list = []
for key in range(len(script_info)):
    trans_dic = okt.nouns(script_info[key])

    voice_dic = okt.nouns(answer_stt[key])

    dic_sum = trans_dic + voice_dic
    feats = set(dic_sum)
    
    trans_arr = np.array(make_matrix(feats, trans_dic))
    voice_arr = np.array(make_matrix(feats, voice_dic))
    
    cos_con = cos_sim(trans_arr, voice_arr) # 코사인 유사도
    i += 1
    print(i,'번째 문장 일치도:',cos_con)
    if np.isnan(cos_con):
      continue
    else:
      cos_list.append(cos_con)

    
```

```python
cos_avg = sum(cos_list)/len(cos_list)
print(f'총 유사도 평균:  {cos_avg:.3f}')
```

## 2)기본 라이브러리(시퀀스매쳐)

```python
i = 0
sm_list = []
for key in range(len(script_info)):
    i+=1
    print('<', i, '>')
    print('정답:', script_info[key])
    print('대사:', answer_stt[key])
    ratio_sm = SequenceMatcher(None, script_info[key], answer_stt[key]).ratio()
    print('유사도:',ratio_sm)
    print('----------------------------------------')
    sm_list.append(ratio_sm)
```

```python
sm_avg = sum(sm_list)/len(sm_list)
print(f'총 유사도 평균:  {sm_avg:.3f}')
```
